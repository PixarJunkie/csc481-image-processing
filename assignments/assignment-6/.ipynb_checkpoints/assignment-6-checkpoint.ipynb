## Problem 1: Erosion and Dilation (20)  
Choose a gray scale image (or convert a color image to grayscale). Convert to binary based on a
mean-intensity threshold. Apply erosion twice and then dilation twice with a structuring element
of your choice. (Defaults are fine.) Then, starting with the original binary image, dilate twice
and erode twice. Show the two resulting images and contrast them. 

```python
#Convert to binary 
thresh = np.mean(gray_img)

#Binary image
binary_img = cv2.threshold(gray_img, thresh, 255, cv2.THRESH_BINARY)[1]

#Plot image
fig = plt.figure()
plt.imshow(binary_img)
fig.set_size_inches(10, 10)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Erode > Dialate binary_image 

#Define kernel
kernel = np.ones((5, 5), np.uint8)

#Erode image
e_d_img = cv2.erode(binary_img, kernel,iterations = 2)
#Dilate image
e_d_img = cv2.dilate(e_d_img, kernel, iterations = 2)
```

```python
#Dilate > Erode binary_image 

#Define kernel
kernel = np.ones((5, 5), np.uint8)

#Dilate image
d_e_img = cv2.dilate(binary_img, kernel, iterations = 2)
#Erode image
d_e_img = cv2.erode(d_e_img, kernel,iterations = 2)
```

```python
#Plot images
plt.clf()
fig = plt.figure()
fig.set_size_inches(20, 20)
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(e_d_img)
ax1.set_title('Erosion > Dilate')
plt.xticks([]), plt.yticks([])
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(d_e_img)
ax2.set_title('Dilate > Erosion')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Thoughts  
Applying erosion before dilation captures darker regions a lot more effectively than dilating prior to erosion. This can be seen in both Mickey and Minnie's eyes, and sweaters. The texture of their faces looks a little more "plastic-y" in the dilate > erosion image. Something that's interesting is what the processes did to the snowflakes on their clothes. Erosion > dilation removed some of the detail in the snowflakes, which is especially noticable in Mickey's sweater. Dilation > erosion kind of blurred them, making them less rigid and it seems to capture edges of dark on dark objects better (Mickey's right arm).

## Problem 2: Opening and Closing (20)
Using the binary image from Problem 1, apply opening twice and then closing twice with a
structuring element of your choice. (Defaults are fine.) Then, starting with the original binary
image, close twice and open twice. Show the two resulting images and contrast them. Then,
compare how your two images from Problem 2 visually differ from the two images of Problem
1.

```python
#Opening > Closing binary_image 

#Define kernel
kernel = np.ones((5, 5), np.uint8)

#Opening image x2
o_c_img = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel)
o_c_img = cv2.morphologyEx(o_c_img, cv2.MORPH_OPEN, kernel)
#Dilate image x2
o_c_img = cv2.morphologyEx(o_c_img, cv2.MORPH_CLOSE, kernel)
o_c_img = cv2.morphologyEx(o_c_img, cv2.MORPH_CLOSE, kernel)
```

```python
#Closing > Opening binary_image 

#Define kernel
kernel = np.ones((5, 5), np.uint8)

#Closing image x2
c_o_img = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)
c_o_img = cv2.morphologyEx(c_o_img, cv2.MORPH_CLOSE, kernel)
#Opening image x2
c_o_img = cv2.morphologyEx(c_o_img, cv2.MORPH_OPEN, kernel)
c_o_img = cv2.morphologyEx(c_o_img, cv2.MORPH_OPEN, kernel)
```

```python
# Plot images
plt.clf()
fig = plt.figure()
fig.set_size_inches(20, 20)
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(o_c_img)
ax1.set_title('Open > Close')
plt.xticks([]), plt.yticks([])
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(c_o_img)
ax2.set_title('Close > Open')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Thoughts  
There are some subtle differences in these two images as compared to problem 1. Opening and then closing still captures some of the dark regions better than closing and then opening. The edge of Mickey's arm is a little more noticable as well, and the snowflakes didn't lose as much definition as they did when eroding > dilating. Closing and then opening captured slightly more dark regions than dilating > eroding. The snowflakes are also not as overly exaggerated.

## Problem 3: Boundary Extraction (10)
Boundary extraction of an image I, β(I), can be performed by eroding I with a structuring
element B and then subtracting the eroded image from I:
β(I) = I – (I ɵ B)
Using your binary image from the two problems above, extract the boundaries. Then, extract
Canny edges from the original grayscale image. Show the two images and compare them.

```python
#Generate structured kernel 
struct_kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

#Erode binary image
eroded_img = cv2.erode(binary_img, struct_kernel, iterations = 1)
#I - B(I)
eroded_edges = binary_img - eroded_img

#Canny edges
canny_edges = cv2.Canny(gray_img, 100, 200)
```

```python
#Plot images
plt.clf()
fig = plt.figure()
fig.set_size_inches(20, 20)
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(eroded_edges)
ax1.set_title('Boundary Extraction')
plt.xticks([]), plt.yticks([])
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(canny_edges)
ax2.set_title('Canny Edges')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Thoughts  
Overall, the boundary extraction method captured the edges better than canny edges in the background, snowflakes, facial details, and arms/hands. Canny edges was able to pick up Mickey's ears, both of their hats, and the bottom of Mickey's sweater better than the boundary extraction method.
