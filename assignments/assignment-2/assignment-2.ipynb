```python
#Resize image
def scale(img, scale):
    img_height = img.shape[0] 
    img_width = img.shape[1]
    img_channels = img.shape[2]
    
    #Downsize image
    if scale < 0:
        new_height = round(img_height / abs(scale))
        new_width = round(img_width / abs(scale))
    
    #Upscale image
    if scale > 0: 
        new_height = img_height * abs(scale)
        new_width = img_width * abs(scale)
    
    #Scaled image
    scaled_height = [int(img_height * y / new_height) for y in range(new_height)]
    scaled_width = [int(img_width * x / new_width) for x in range(new_width)]
    
    scaled_img =  [[img[int(img_height * y / new_height)] [int(img_width * x / new_width)] for x in range(new_width)] for y in range(new_height)]

    return scaled_img
```

### Contrast Stretching

```python
#Image histogram
plt.hist(img.ravel(), 256, [0, 256])
```

```python
#Equalizing RGB

r, g, b = cv2.split(img)
eq_r = cv2.equalizeHist(r) #equalizing red component
eq_g = cv2.equalizeHist(g) #equalizing green component
eq_b = cv2.equalizeHist(b) #equalizing blue component

#Merge image
eq_img = cv2.merge((eq_r, eq_g, eq_b)) 

#Plot new histogram
plt.hist(eq_img.ravel(), 256, [0, 256])
```

```python
plt.imshow(img)
plt.title('Original')
```
