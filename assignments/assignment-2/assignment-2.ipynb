Problem 1: Zooming/Oversampling and Shrinking/Subsampling Images by Pixel Replication(15/15)Write your own function capable of shrinking and zooming an image by pixel replication and decimation. Assume that the desired zoom/shrink factors will be the inputs for your function and will have integer values: a negative input means shrink and a positive input means expand. Do not use MatLab built in functions (or other language libraries) for decimation and replication, although you can use other MatLab functions.  
(a) Use your program to shrink an image by a factor of 4 in each dimension. Show the shrunk image.

```python
#Resize image
def scale(img, scale):
    img_height = img.shape[0] 
    img_width = img.shape[1]
    img_channels = img.shape[2]
    
    #Downsize image
    if scale < 0:
        new_height = round(img_height / abs(scale))
        new_width = round(img_width / abs(scale))
    
    #Upscale image
    if scale > 0: 
        new_height = img_height * abs(scale)
        new_width = img_width * abs(scale)
    
    #Scaled image
    scaled_height = [int(img_height * y / new_height) for y in range(new_height)]
    scaled_width = [int(img_width * x / new_width) for x in range(new_width)]
    
    scaled_img =  [[img[int(img_height * y / new_height)] [int(img_width * x / new_width)] for x in range(new_width)] for y in range(new_height)]

    return scaled_img
```
